## Using the library

````python
from quantsim.core import Qubit
````

````python
myqubit = Qubit()

print(myqubit)
````

Now that we have a qubit lets perform an observation

````python
myqubit.observe()
````

The qubit has collapsed into a bit

Using gates

````python
from quantsim.gates import Identity, PauliX

myqubit = Qubit()
identity = Identity()
pauliX = PauliX()
````

The Identity gate should not make any changes to our qubit

````python
print(myqubit)

print(Qubit(matrix=(identity*myqubit)))
print(Qubit(matrix=(identity*myqubit)))
````

````python
from quantsim.circuit import *
from quantsim.gates import PauliX


myq = QubitComponent([])
myg  = GateComponent(PauliX(), [], [])
myo = Observer([])

myq.add_output(myg)
myg.add_output(myo)

myc = Circuit()
myc.add(myq)
myc.add(myg)
myc.add(myo)
````

````python
myc.calculate()
````

````python
print(myo.value)
````
