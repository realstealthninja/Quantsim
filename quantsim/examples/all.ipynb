# Quantsim
Welcome to quantsim, A simple python library for performing quantum linear algebra.
This note book is meant to be both an introduction as well as tutorial to the library as a whole

## Installing the library

Installation can be done via the **PIP**


## Using the library

````python
from quantsim.core import Qubit
````

````python
myqubit = Qubit()

print(myqubit)
````

Now that we have a qubit lets perform an observation

````python
myqubit.observe()
````

The qubit has collapsed into a bit

Using gates

````python
from quantsim.gates import Identity, PauliX

myqubit = Qubit()
identity = Identity()
pauliX = PauliX()
````

The Identity gate should not make any changes to our qubit

````python
print(myqubit)

print(Qubit(matrix= (identity*myqubit)))
````
